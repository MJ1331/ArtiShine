```python
from PIL import Image, ImageDraw, ImageFont
import datetime
import matplotlib.pyplot as plt

# Data
data = {
    'name': "Sita Sharma",
    'shop_name': "Clay Creations",
    'location': "Jaipur, India",
    'dob': datetime.date(1995, 7, 21),
}

# Load your image from disk
image_path = r"C:\Users\Sheraz\Documents\pythontest\project_all_ml\GoogleHackathon\Onboarding_Post\ARTISHIINE.png"
image = Image.open(image_path).convert('RGB')
draw = ImageDraw.Draw(image)

# Load font (fallback if arial is missing)
font_path = r"C:\Users\Sheraz\Documents\pythontest\project_all_ml\GoogleHackathon\Onboarding_Post\Pixel Game.otf"   # Specify a valid TrueType font file path here

# Text lines
# Each tuple = (text, font_size, extra_spacing_after)
lines = [
    ("INTRODUCING", 250, 0),       # Add more space *after* this line
    (f"{data['name']}", 180, 80),                      # spacer line
    (f"Shop : {data['shop_name']}", 120, 50),
    (f"Location : {data['location']}", 120, 50),
    (f"DOB : {data['dob'].strftime('%d-%m-%Y')}", 120, 0),
]

W, H = image.size
start_y = 250
y=start_y

for (line, size, extra_space) in lines:
    try:
        font = ImageFont.truetype(font_path, size)
    except:
        font = ImageFont.load_default()
    
    text_width, text_height = draw.textsize(line, font=font)
    x = W / 2
    draw.text((x, y), line, font=font, fill=(0,0,0), anchor="mm")
    y+=text_height+extra_space

# Display in notebook
plt.imshow(image)
plt.axis('off')
plt.show()
```
